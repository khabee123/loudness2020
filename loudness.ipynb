```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('spotify.csv')

df2 = df.sort_values(by=['popularity'])
df2 = df2.reset_index(drop=True)
```

```python
plt.hist(df2['loudness'])
plt.axis([-20, 0,0,40])
plt.ylabel('Amount of songs')
plt.xlabel('Loudness (db)')
plt.suptitle('Loudness')
plt.show()
```

```python
plt.scatter(df2['popularity'],df2['loudness'])
plt.ylabel('Loudness (db)')
plt.xlabel('Popularity')
plt.suptitle('Popularity and Loudness')
plt.show()
```
